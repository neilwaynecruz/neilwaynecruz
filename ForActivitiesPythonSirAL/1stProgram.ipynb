```python
import time

# for writing

import csv
from os import path
import time

try:
    # Determine file mode
    if path.exists("attendance.csv"):
        mode = "a"
    else:
        mode = "w"

    # Open the file
    with open("attendance.csv", mode, newline="") as fp:
        fprint = csv.writer(fp)
        records = []

        if mode == "w":
            fprint.writerow(["ID", "Last Name","First Name","Hours","  Status","\tTime Stamp"])

        # Collect employee data
        while True:
            emp_id = input("Enter 'Employee ID': ")
            emp_Lname = input("Enter employee last name: ")
            emp_Fname = input("Enter employee first name: ")
            emp_Hwork = int(input("Enter number of working hours: "))
            emp_status = ""

            # Determine employment status
            if emp_Hwork >= 40:
                emp_status = "Full Time"
            elif emp_Hwork >= 20:
                emp_status = "Part Time"
            else:
                emp_status = "Casual"

            emp_time = time.strftime("%D:%H:%M:%S",time.localtime())

            # Add record to the list
            temp_list = [emp_id, emp_Lname, emp_Fname, emp_Hwork, emp_status,emp_time]
            records.append(temp_list)

            # Ask if the user wants to continue
            choice = input("Continue [Y/N]? ").strip().upper()
            if not choice or choice[0] == "N":
                break
        
        fprint.writerows(records)
        print("Record has been added.")

except FileNotFoundError as e:
    print(f"Error: {e}")
except IsADirectoryError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error: {e}")
finally:
    print("Program Ended")
```

```python
class People:
    def __init__(self,name,age,height,weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg.")

    def birthDay(self,age):
        self.age += 1
        print(f"Happy birthday {self.name}, you are now {self.age} years old!")

class Student(People):
    def __init__(self,name,age,height,weight,studentID):
        super().__init__(name,age,height,weight)
        self.studentID = studentID

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg. My student ID is {self.studentID}.")

class Teacher(People):
    def __init__(self,name,age,height,weight,employeeID):
        super().__init__(name,age,height,weight)
        self.employeeID = employeeID

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg. My employee ID is {self.employeeID}.")

people = [Student("Alice", 20, 170, 60, "S123"), 
          Teacher("Bob", 40, 180, 80, "T456")]

for person in people:
    person.introduce()
    person.birthDay(person.age)
    print()
```

```python
def is_pangram(sentence):
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    return alphabet.issubset(set(sentence.lower()))

# Example Usage
sentence = input("Enter a sentence: ")
if is_pangram(sentence):
    print("The sentence is a pangram.")
else:
    print("The sentence is NOT a pangram.")
```
