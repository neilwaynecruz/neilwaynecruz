```python
from tkinter import *
import tkinter.messagebox as tm

my_window = Tk()
my_window.title('Quarter Grades Calculator')

# Functions
def compute_average(): # eto method na to is para sa computation lang ng ave 
    try:
        q1 = float(txtQ1.get()) # get mo yung txtQ1 then type cast(convert) mo siya sa float
        q2 = float(txtQ2.get()) # get mo yung txtQ2 then type cast(convert) mo siya sa float
        q3 = float(txtQ3.get()) # get mo yung txtQ3 then type cast(convert) mo siya sa float
        q4 = float(txtQ4.get()) # get mo yung txtQ4 then type cast(convert) mo siya sa float
        system = grading_system.get() #get mo lang kung ano yung selected na radiobutton
        print(system)


        # condition kung ano yung selected na radio button yun yung average computation na gagawin niyan (gets baa????)
        if system == "Average": 
            ave = (q1 + q2 + q3 + q4) / 4
        elif system == "Weighted":
            ave = (q1*0.2 + q2*0.2 + q3*0.2 + q4*0.4)
        elif system == "Cumulative":
            ave = q4
        else:
            ave = 0
        txtAve.set(round(ave, 2)) # set mo yung average na nakuha mo sa variable na 'txtAve'
    except Exception as e: # if my exception na nangyari sa function na to dito pupunta. Alam niyo na yan AHAHAHAHA
        tm.showerror('Error', 'Please enter valid numbers for all quarters.')

def on_clear(): #clear mo yung mga textfield
    txtQ1.set('')
    txtQ2.set('')
    txtQ3.set('')
    txtQ4.set('')
    txtAve.set('')
    grading_system.set(0) # once yung clear button is na click babalik yung selected radiobutton sa Average radio button / pwede niyo rin naman tanggalin to (BAHALA NA KAYO HAHAHA)

def on_close(): # message box/dialog lang to guys
    ans = tm.askyesno('Close', 'Are you sure you want to close the application?')
    if ans:
        my_window.destroy()

# Initialization ng Variables
txtQ1 = StringVar()
txtQ2 = StringVar()
txtQ3 = StringVar()
txtQ4 = StringVar()
txtAve = StringVar()
grading_system = StringVar()
grading_system.set(0)

# Label at entry
Label(my_window, text='1st Quarter').grid(row=0, column=0, padx=10, pady=5)
Entry(my_window, textvariable=txtQ1).grid(row=0, column=1, padx=10, pady=5)

Label(my_window, text='2nd Quarter').grid(row=1, column=0, padx=10, pady=5)
Entry(my_window, textvariable=txtQ2).grid(row=1, column=1, padx=10, pady=5)

Label(my_window, text='3rd Quarter').grid(row=2, column=0, padx=10, pady=5)
Entry(my_window, textvariable=txtQ3).grid(row=2, column=1, padx=10, pady=5)

Label(my_window, text='4th Quarter').grid(row=3, column=0, padx=10, pady=5)
Entry(my_window, textvariable=txtQ4).grid(row=3, column=1, padx=10, pady=5)

Label(my_window, text='Average').grid(row=4, column=0, padx=10, pady=5)
Entry(my_window, textvariable=txtAve, state='readonly').grid(row=4, column=1, padx=10, pady=5)

Label(my_window, text='Grading System').grid(row=5, column=0, columnspan=1)

#Radio Button
Radiobutton(my_window, text='Averaging', variable=grading_system, value='Average').grid(row=6, column=0, columnspan=1, padx=10)
Radiobutton(my_window, text='Weighted Averaging', variable=grading_system, value='Weighted').grid(row=7, column=0, columnspan=1, padx=10)
Label(my_window, text="").grid(row=7, column=1) 
Radiobutton(my_window, text='Cumulative', variable=grading_system, value='Cumulative').grid(row=8, column=0, columnspan=1, padx=10)

# para sa button
Button(my_window, text='OK', command=compute_average, width=10).grid(row=9, column=0, padx=10, pady=15)
Button(my_window, text='CLEAR', command=on_clear, width=10).grid(row=9, column=1, padx=10, pady=15)
Button(my_window, text='CLOSE', command=on_close, width=10).grid(row=9, column=2, padx=10, pady=15)

my_window.mainloop()
```

# ODD AND EVEN

```python
from tkinter import *

def check_numbers():
    try:
        # Convert inputs to integers
        values = [int(var.get()) for var in inputs]

        # Clear output
        for var in odd_results + even_results:
            var.set("")

        # Filter and show based on choice
        if choice.get() == 1:  # ODD
            odd = [str(n) for n in values if n % 2 != 0]
            for i, val in enumerate(odd):
                odd_results[i].set(val)
        elif choice.get() == 2:  # EVEN
            even = [str(n) for n in values if n % 2 == 0]
            for i, val in enumerate(even):
                even_results[i].set(val)
    except ValueError:
        print("Invalid input!")

def clear_fields():
    for var in inputs + odd_results + even_results:
        var.set("")
    choice.set(0)

# Main window
win = Tk()
win.title("Odd or Even Checker")

# Variables
inputs = [StringVar() for _ in range(10)]
odd_results = [StringVar() for _ in range(10)]
even_results = [StringVar() for _ in range(10)]
choice = IntVar()

# Title
Label(win, text="Enter 10 Numbers", font=("Arial", 15, "bold")).grid(row=0, column=0, columnspan=2)

# Input fields
for i, var in enumerate(inputs):
    Entry(win, textvariable=var, width=10, borderwidth=5).grid(row=1 + i // 2, column=i % 2)

# Labels
Label(win, text="ODD", font=("Arial", 15, "bold")).grid(row=7, column=0)
Label(win, text="EVEN", font=("Arial", 15, "bold")).grid(row=7, column=1)

# Output fields
for i in range(10):
    Entry(win, textvariable=odd_results[i], width=10, borderwidth=5, state='readonly').grid(row=8 + i, column=0)
    Entry(win, textvariable=even_results[i], width=10, borderwidth=5, state='readonly').grid(row=8 + i, column=1)

# Radio buttons
Radiobutton(win, text="ODD", variable=choice, value=1, command=check_numbers, font=("Arial", 14)).grid(row=18, column=0)
Radiobutton(win, text="EVEN", variable=choice, value=2, command=check_numbers, font=("Arial", 14)).grid(row=19, column=0)

# Buttons
Button(win, text="Clear", command=clear_fields, width=10, font=("Arial", 12)).grid(row=20, column=0, pady=10)
Button(win, text="Close", command=win.destroy, width=10, font=("Arial", 12)).grid(row=20, column=1, pady=10)

# Run the app
win.mainloop()
```

```python
from tkinter import *
import tkinter.messagebox as tm

my_window = Tk()
my_window.title('Transmuted Grade Activity')

def transmute_grade(tg):
    if 97 <= tg <= 100:
        return '1.00'
    elif 94 <= tg <= 96:
        return '1.25'
    elif 91 <= tg <= 93:
        return '1.50'
    elif 88 <= tg <= 90:
        return '1.75'
    elif 85 <= tg <= 87:
        return '2.00'
    elif 82 <= tg <= 84:
        return '2.25'
    elif 79 <= tg <= 81:
        return '2.50'
    elif 76 <= tg <= 78:
        return '2.75'
    elif tg == 75:
        return '3.00'
    elif tg == 74:
        return '3.25'
    elif 72 <= tg <= 73:
        return '3.50'
    elif 70 <= tg <= 71:
        return '3.75'
    elif 65 <= tg <= 69:
        return '4.00'
    elif 60 <= tg <= 64:
        return '4.25'
    elif 58 <= tg <= 59:
        return '4.50'
    elif 50 <= tg <= 57:
        return '5.00'
    else:
        return 'Invalid'

def on_Ok():
    try:
        score = int(txtScore.get())
        items = int(txtItems.get())
        tg = score * 50 / items + 50
        txtTG.set(round(tg, 2))
        txtEQ.set(transmute_grade(tg))

        score2 = int(txtScore2.get())
        items2 = int(txtItems2.get())
        tg2 = score2 * 50 / items2 + 50
        txtTG2.set(round(tg2, 2))
        txtEQ2.set(transmute_grade(tg2))

        eq = float(txtEQ.get())
        eq2 = float(txtEQ2.get())
        
        ave = (eq + eq2) / 2
        txtAve.set(round(ave, 2))

        percent = ave * 0.35
        txtPercent.set(round(percent,2))

    except Exception as e:
        tm.showerror('Error', str(e))

def on_clear():
    txtScore.set('')
    txtItems.set('')
    txtTG.set('')
    txtEQ.set('')
    txtScore2.set('')
    txtItems2.set('')
    txtTG2.set('')
    txtEQ2.set('')
    txtAve.set('')
    txtPercent.set('')

def on_close():
    ans = tm.askyesno('Close', 'Weh?')
    if ans:
        my_window.destroy()

txtScore = StringVar()
txtItems = StringVar()
txtTG = StringVar()
txtEQ = StringVar()
txtScore2 = StringVar()
txtItems2 = StringVar()
txtTG2 = StringVar()
txtEQ2 = StringVar()
txtAve = StringVar()
txtPercent = StringVar()


Label(my_window, text='Quiz 1').grid(row=0, column=0)
Label(my_window, text='Enter Score:').grid(row=1, column=0)
Entry(my_window, textvariable=txtScore).grid(row=1, column=1)
Label(my_window, text='Enter no. of items:').grid(row=2, column=0)
Entry(my_window, textvariable=txtItems).grid(row=2, column=1)
Label(my_window, text='TG:').grid(row=3, column=0)
Entry(my_window, textvariable=txtTG,state = 'disabled').grid(row=3, column=1)
Label(my_window, text='EQ:').grid(row=4, column=0)
Entry(my_window, textvariable=txtEQ,state = 'disabled').grid(row=4, column=1)

Label(my_window, text='Quiz 2').grid(row = 5, column=0)
Label(my_window, text='Enter Score:').grid(row = 6, column=0)
Entry(my_window, textvariable=txtScore2).grid(row = 6, column = 1)
Label(my_window, text='Enter no. of items:').grid(row=7, column=0)
Entry(my_window, textvariable=txtItems2).grid(row=7, column=1)
Label(my_window, text='TG:').grid(row=8, column=0)
Entry(my_window, textvariable=txtTG2,state = 'disabled').grid(row=8, column=1)
Label(my_window, text='EQ:').grid(row=9, column=0)
Entry(my_window, textvariable=txtEQ2,state = 'disabled').grid(row=9, column=1)


Label(my_window, text='Quiz Average:').grid(row=10, column=0)
Entry(my_window, textvariable=txtAve,state = 'disabled').grid(row=10, column=1)
Label(my_window, text = '35% ').grid(row = 11, column = 0)
Entry(my_window, textvariable = txtPercent, state = 'disabled',width = 15, border = 2).grid(row = 11, column = 1)

Button(my_window, text='OK', command=on_Ok, width=10).grid(row = 12, column=0)
Button(my_window, text='Clear', command=on_clear, width=10).grid(row = 12, column=1)
Button(my_window, text='Close', command=on_close, width=10).grid(row = 12, column=2)

my_window.mainloop()

```

```python
from tkinter import *
import math

# Create the main window
my_window = Tk()
my_window.title("Calculator")

# Variable to store the result
result = StringVar()

# Entry widget for input/result
Entry(my_window, textvariable=result, font=("Arial", 24), justify='right',state='readonly').grid(row=0, column=0, columnspan=4, pady=10)

# Add a single label for memory buttons
Label(my_window, text="MC   MR    M+    M-    MS", font=("Arial", 16)).grid(row=1, column=0, columnspan=4, pady=5)

# Button click handler
def button_click(value):
    try:
        if value == "=":
            result.set(eval(result.get()))
        elif value == "C":
            result.set("")
        elif value == "²√x":
            result.set(math.sqrt(float(result.get())))
        elif value == "x²":
            result.set(float(result.get()) ** 2)
        elif value == "1/x":
            result.set(1 / float(result.get()))
        else:
            result.set(result.get() + value)
    except:
        result.set("Error")

# Button layout
buttons = [
    ('%', 2, 0), ('CE', 2, 1), ('C', 2, 2), ('X', 2, 3),
    ('1/x', 3, 0), ('x²', 3, 1), ('²√x', 3, 2), ('÷', 3, 3),
    ('7', 4, 0), ('8', 4, 1), ('9', 4, 2), ('x', 4, 3),
    ('4', 5, 0), ('5', 5, 1), ('6', 5, 2), ('-', 5, 3),
    ('1', 6, 0), ('2', 6, 1), ('3', 6, 2), ('+', 6, 3),
    ('C', 7, 0), ('0', 7, 1), ('.', 7, 2), ('=', 7, 3),
]

# Add buttons to the grid
for text, row, col in buttons:
    Button(my_window, text=text, width=8, height=3, command=lambda t=text: button_click(t)).grid(row=row, column=col, padx=5, pady=5)

# Run the main loop
my_window.mainloop()
```

```python
import time

# for writing

import csv
from os import path
import time

try:
    # Determine file mode
    if path.exists("attendance.csv"):
        mode = "a"
    else:
        mode = "w"

    # Open the file
    with open("attendance.csv", mode, newline="") as fp:
        fprint = csv.writer(fp)
        records = []

        if mode == "w":
            fprint.writerow(["ID", "Last Name","First Name","Hours","  Status","\tTime Stamp"])

        # Collect employee data
        while True:
            emp_id = input("Enter 'Employee ID': ")
            emp_Lname = input("Enter employee last name: ")
            emp_Fname = input("Enter employee first name: ")
            emp_Hwork = int(input("Enter number of working hours: "))
            emp_status = ""

            # Determine employment status
            if emp_Hwork >= 40:
                emp_status = "Full Time"
            elif emp_Hwork >= 20:
                emp_status = "Part Time"
            else:
                emp_status = "Casual"

            emp_time = time.strftime("%D:%H:%M:%S",time.localtime())

            # Add record to the list
            temp_list = [emp_id, emp_Lname, emp_Fname, emp_Hwork, emp_status,emp_time]
            records.append(temp_list)

            # Ask if the user wants to continue
            choice = input("Continue [Y/N]? ").strip().upper()
            if not choice or choice[0] == "N":
                break
        
        fprint.writerows(records)
        print("Record has been added.")

except FileNotFoundError as e:
    print(f"Error: {e}")
except IsADirectoryError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error: {e}")
finally:
    print("Program Ended")
```

```python
class People:
    def __init__(self,name,age,height,weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg.")

    def birthDay(self,age):
        self.age += 1
        print(f"Happy birthday {self.name}, you are now {self.age} years old!")

class Student(People):
    def __init__(self,name,age,height,weight,studentID):
        super().__init__(name,age,height,weight)
        self.studentID = studentID

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg. My student ID is {self.studentID}.")

class Teacher(People):
    def __init__(self,name,age,height,weight,employeeID):
        super().__init__(name,age,height,weight)
        self.employeeID = employeeID

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, my height is {self.height} cm and my weight is {self.weight} kg. My employee ID is {self.employeeID}.")

people = [Student("Alice", 20, 170, 60, "S123"), 
          Teacher("Bob", 40, 180, 80, "T456")]

for person in people:
    person.introduce()
    person.birthDay(person.age)
    print()
```

```python
def is_pangram(sentence):
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    return alphabet.issubset(set(sentence.lower()))

# Example Usage
sentence = input("Enter a sentence: ")
if is_pangram(sentence):
    print("The sentence is a pangram.")
else:
    print("The sentence is NOT a pangram.")
```
